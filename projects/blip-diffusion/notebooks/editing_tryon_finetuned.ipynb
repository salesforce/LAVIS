```python
import torch

from PIL import Image
from lavis.models import load_model_and_preprocess
```

```python
torch.cuda.is_available()
```

```python
model, vis_preprocess, txt_preprocess = load_model_and_preprocess("blip_diffusion", "base", device="cuda", is_eval=True)
```

### Description
This demo shows how to edit a **real** image, in a virtual try-on application with a fine-tuned checkpoint. It works in the following steps:

(1) Load the fine-tuned checkpoint.

(2) run DDIM inversion on the ``src_image`` using prompt ``A ${src_subject} ${prompt}.``; the ``src_subject`` can be a noun for the outfit, e.g. dress, shirt, jacket, depending on the ``src_image``.

(3) extracting BLIP-2 embeddings on condition subject image, using ``cond_subject`` and ``cond_image``. The condition image can be a different outfit/clothes to try on.

(4) edit the real image with the subject visuals, using the prompt ``A ${BLIP-2 embedding} ${tgt_subject} ${prompt}`` and the DDIM inverted latents.

### Tips

(1) Best results are obtained if the shapes of the clothing in the conditional image and the source image are similar.

(2) Try different ``lb_threshold`` to adjust the area of the edited regions.

```python
# finetuned_ckpt = "/export/home/workspace/LAVIS-Diffusion/LAVIS/lavis/output/debug/BLIP-diffusion/finetune/shein-jacket/20230622075/checkpoint_120.pth"
finetuned_ckpt = "/export/share/dongxuli/zerobooth/shein-jacket/checkpoint_120.pth"
model.load_checkpoint(finetuned_ckpt)
```

```python
cond_subject = "jacket"
src_subject = "jacket"
tgt_subject = "jacket"

text_prompt = "on a person, flower field in the background"

src_subject = txt_preprocess["eval"](src_subject)
tgt_subject = txt_preprocess["eval"](tgt_subject)
cond_subject = txt_preprocess["eval"](cond_subject)
text_prompt = [txt_preprocess["eval"](text_prompt)]

src_image = Image.open("../images/junnan-profile.jpg").convert("RGB")

display(src_image.resize((256, 256)))
```

```python
samples = {
    "cond_images": None,
    "cond_subject": cond_subject,
    "src_subject": src_subject,
    "tgt_subject": tgt_subject,
    "prompt": text_prompt,
    "raw_image": src_image,
}
```

```python
iter_seed = 8887
guidance_scale = 7.5
num_inference_steps = 50
num_inversion_steps = 50
lb_threshold = 0.45
negative_prompt = "over-exposure, under-exposure, saturated, duplicate, out of frame, lowres, cropped, worst quality, low quality, jpeg artifacts, morbid, mutilated, out of frame, ugly, bad anatomy, bad proportions, deformed, blurry, duplicate"

output = model.edit(
    samples,
    seed=iter_seed,
    guidance_scale=guidance_scale,
    num_inference_steps=num_inference_steps,
    num_inversion_steps=num_inversion_steps,
    neg_prompt=negative_prompt,
    lb_threshold=lb_threshold
)

print("=" * 30)
print("Before editing:")
display(output[0])

print("After editing:")
display(output[1])
```
